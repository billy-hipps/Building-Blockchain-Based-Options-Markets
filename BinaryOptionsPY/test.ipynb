```python
import yfinance as yf

def get_live_stock_price_yfinance(ticker):
    stock = yf.Ticker(ticker)
    stock_info = stock.info
    
    market_price = stock_info.get("regularMarketPrice")  # Direct live price
    if market_price is None:
        return f"âš  No live price found for {ticker}."
    
    return f"Live {ticker} Stock Price: ${market_price}"

print(get_live_stock_price_yfinance("AAPL"))

```
